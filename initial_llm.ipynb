```python
import torch
from transformers import LlamaTokenizer, LlamaForCausalLM
```

```python
model = 'openlm-research/open_llama_3b'
```

```python
tokenizer = LlamaTokenizer.from_pretrained(model)
```

```python
base_model = LlamaForCausalLM.from_pretrained(model, torch_dtype=torch.float16, device_map='auto')
```

```python
instruction = 'What is the difference between World War 1 and World War 2?'
```

```python
input_ids = tokenizer(instruction, return_tensors = "pt").input_ids.cuda() # tokenization step
```

```python
generated_output = base_model.generate(input_ids = input_ids, max_new_tokens = 64) # generation step
```

```python
print(tokenizer.decode(generated_output[0]))
```

```python
instruction2 = 'What is the difference between whales and sharks?'
input_ids2 = tokenizer(instruction2, return_tensors = "pt").input_ids.cuda() # tokenization step
generated_output2 = base_model.generate(input_ids = input_ids2, max_new_tokens = 64) # generation step
print(tokenizer.decode(generated_output2[0]))
```
